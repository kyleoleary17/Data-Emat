1. To start off, I want to see the amount of players in each position that have had an injuries recorded. I used matplotlib using plt and seaborn using sns to create a graph. Based off these results, we can see that the cornerback position has the most amount of injuries reported and the long snapper has the least. Looking at the results, it is clear that the positions that are the most physicla have the most reported injuries, and the positions with less physical contact the amount of injuries go down

```python
ax1 = ad['Position'].value_counts().head(10).index
plt.figure(figsize=(10, 5))
sns.countplot(data=ad, y='Position')
plt.title('NFL Positions')
plt.ylabel('NFL Positions')
plt.xlabel('Count')
```

1. Next, I wanted to see the injury types that were recorded. Based off these results, we can see over 300 of them were concussions, about 65 were head, and very few were illness. Head and concussions kind of go hand in hand, they both involve resulting in contact to the head, but they are different because a head injury is an injury to the head, and a concussion is an injury to the brain.

```python
ax2 = ad['Reported Injury Type'].value_counts().head(10).index
plt.figure(figsize=(10, 5))
sns.countplot(data=ad, y='Reported Injury Type')
plt.title('NFL Reported Injury Type')
plt.ylabel('Reported Injury Type')
plt.xlabel('Count')
```

1. After finding out the types of injuries that were recorded, I wanted to see if there were any injuries that were unknown. Based off these results, most were known, about 380, but there were a sliver of injuries that were unknown, about 10.

```python
ax3 = ad['Unknown Injury?'].value_counts().head(10).index
plt.figure(figsize=(10, 5))
sns.countplot(data=ad, y='Unknown Injury?')
plt.title('NFL Unknown Injuries?')
plt.ylabel('Unknown Injury?')
plt.xlabel('Count')
```

1. Now that I have gathered all the information I need about the injuries, I wanted to see how many injuries took place in the pre-season. According to these results, there werent a lot of injuries that took place in the pre-season, but there were some that did, about 30. This was to be excpected because the pre-season is a shorter amount of games than the regular season.

```python
ax3 = ad['Pre-Season Injury?'].value_counts().head(10).index
plt.figure(figsize=(10, 5))
sns.countplot(data=ad, y='Pre-Season Injury?')
plt.title('NFL Pre-Season Injury?')
plt.ylabel('Pre-Season Injury?')
plt.xlabel('Count')
```

1. After gathering the information about the pre-season injuries, I wanted to see overall how many weeks these players were injured. Accordint to this graph, more than 350 players were injured for 1 week or more, and very few were injured for less than 1 week. For the less than 1 week players, I would say that those are most likely the players that have had head injuries or an unknown injury, since those take less time to recover from than a concussion.

```python
ax3 = ad['Weeks Injured'].value_counts().head(10).index
plt.figure(figsize=(10, 5))
sns.countplot(data=ad, y='Weeks Injured')
plt.title('NFL Weeks Injured')
plt.ylabel('Weeks Injured')
plt.xlabel('Count')
```

1. Now that I have seen the amount of weeks injured, I wanted to break that down into seasons. According to this data, the amount of weeks injured actually DECREASED as the seasons went on. I found this information a little shocking. I honestly thought that the amount of weeks injured by season would increase since football is very physical. The only reason I could think of as to why it decreases is because there might have been a new helmet that was created that had more padding and safer for players. Other than that im not sure as to why this could be and this shocked me.

```python
plt.figure(figsize = (10, 5))
ax = sns.countplot(x='Season', hue='Weeks Injured', data=ad)

for p in ax.patches:
    ax.annotate(text=p.get_height(), xy=(p.get_x()+0.2, p.get_height()+200), horizontalalignment='center', verticalalignment='top')

plt.title('NFL Season by Weeks Injured')    
plt.xlabel('NFL Season')
plt.ylabel('Counts')
plt.show()
```
